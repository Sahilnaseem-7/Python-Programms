```python
import matplotlib.pyplot as plt

#Define some data for the plot
x = [2, 6, 7, 9, 10]
y = [7, 5, 6, 3, 11]

#Create the line plot
plt.plot(x,y,  marker='o', linestyle='--', color = 'g', label = 'Data')

#Add labels to the axis 
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

#Add title to the plot
plt.title('A Line Plot')

#Add a legend 
plt.legend

#Show
plt.show()
```

```python
#Drawing a line plot
import matplotlib.pyplot as plt
import numpy as np

#Preparing graph
months = np.arange(1,13)
product_a_sales = np.array([11000, 13000, 15000, 9000, 17000, 14000, 12500, 11500, 18500, 21000, 25000, 34000])
product_b_sales = np.array([8000, 7500, 9500, 10500, 11000, 13500, 15500, 17500, 18000, 21000, 30000, 18000])

#Plotting graph
plt.plot(months, product_a_sales, marker='o', linestyle='-', color = 'y', label = 'Product A')
plt.plot(months, product_b_sales, marker='s', linestyle='--', color = 'r', label = 'Product B')
plt.xlabel('Month')
plt.legend()
plt.ylabel('Sales in a year')
plt.show()
```

```python
#Drawing a Bar graph
import matplotlib.pyplot as plt
import numpy as np

#Preparing graph 
cities = ['Gorakhpur', 'Deoria', 'Kushinagar', 'Azamgarh', 'Balia']
population = [4500000, 3000000, 3500000, 4500000, 3500000]

#Plotting Bar Graph
plt.bar(cities,population, color ='green')
plt.xlabel('City')
plt.ylabel('Population')
plt.title('Population of cities of Uttarpradesh')
plt.show()
```
