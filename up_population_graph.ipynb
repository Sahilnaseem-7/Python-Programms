```python
#Importing neccessary packages
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
import seaborn as sns
```

```python
#Loading dataset
df = pd.read_csv(r'C:/Users/sahil/OneDrive/Documents/up_population.csv')
```

```python
df
```

```python
#Arranging Rows and columnns for the graph
female_population = np.array(df['Female Population'])
male_population = np.array(df['Male Population'])
district = np.array(df['District'])
total_population = np.array(df['Total Population'])
```

```python
#Total Population district-wise
plt.figure(figsize=(15,15))
plt.barh(district, total_population, color='blue')
plt.xlim()
plt.ylabel('Districts of Uttar Pradesh')
plt.xlabel('Population of Districts')
plt.title('Population of UP District-Wise')
plt.show()
```

```python
#Male population district-wise
plt.figure(figsize=(15,15))
plt.barh(district, male_population, color= 'green')
plt.ylabel('Districts of Uttar Pradesh')
plt.xlabel('Male Population ')
plt.title('Male Population of UP District-Wise')
plt.show()

```

```python
#Female population district-wise
plt.figure(figsize=(15,15))
plt.barh(district, female_population, color='red')
plt.ylabel('Districts of Uttar Pradesh')
plt.xlabel('Female Population ')
plt.title('Female Population of UP District-Wise')
plt.show()
```
